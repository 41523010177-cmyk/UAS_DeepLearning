```python
# ===============================
# UAS DEEP LEARNING - FINAL CELL
# ===============================

import os, cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# ====== DATASET ======
# Unzip the dataset
!unzip -q /content/PlantVillage.zip -d /content/

BASE_DIR = "/content/PlantVillage"
IMG_SIZE = 128

images, labels = [], []

for folder in os.listdir(BASE_DIR):
    folder_path = os.path.join(BASE_DIR, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            img = cv2.imread(os.path.join(folder_path, file))
            if img is not None:
                img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                images.append(img)
                labels.append(folder)

images = np.array(images) / 255.0
labels = np.array(labels)

le = LabelEncoder()
labels = le.fit_transform(labels)
num_classes = len(np.unique(labels))

X_train, X_test, y_train, y_test = train_test_split(
    images, labels, test_size=0.2, random_state=42
)

# ====== FUNGSI MODEL ======
def build_model(activation='relu', lr=0.001):
    model = Sequential([
        Conv2D(32, (3,3), activation=activation, input_shape=(IMG_SIZE, IMG_SIZE, 3)),
        MaxPooling2D(2,2),
        Conv2D(64, (3,3), activation=activation),
        MaxPooling2D(2,2),
        Flatten(),
        Dense(128, activation=activation),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(
        optimizer=Adam(lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

    # ====== TABEL ARSITEKTUR MODEL (BASELINE) ======
print("\nTABEL ARSITEKTUR MODEL CNN (BASELINE)")
baseline_model = build_model('relu', 0.001)
baseline_model.summary()


# ====== EKSPERIMEN (P4.1) ======
configs = [
    ('ReLU_lr_0.001', 'relu', 0.001),
    ('ReLU_lr_0.0001', 'relu', 0.0001),
    ('Tanh_lr_0.001', 'tanh', 0.001)
]

results = []
histories = {}

for name, act, lr in configs:
    print(f"\nTraining {name}")
    model = build_model(act, lr)
    history = model.fit(
        X_train, y_train,
        epochs=5,
        batch_size=32,
        validation_split=0.2,
        verbose=1
    )
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    results.append({'Konfigurasi': name, 'Val_Accuracy': acc, 'Val_Loss': loss})
    histories[name] = history

df_results = pd.DataFrame(results)
print("\nTABEL PERBANDINGAN:")
display(df_results)

# ====== GRAFIK PERBANDINGAN ======
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(df_results['Konfigurasi'], df_results['Val_Accuracy'])
plt.title('Akurasi Validasi')

plt.subplot(1,2,2)
plt.bar(df_results['Konfigurasi'], df_results['Val_Loss'])
plt.title('Loss Validasi')
plt.tight_layout()
plt.show()

# ====== KURVA TRAIN vs VALIDATION (P4.2) ======
for name, h in histories.items():
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(h.history['accuracy'], label='Train')
    plt.plot(h.history['val_accuracy'], label='Validation')
    plt.title(f'Accuracy - {name}')
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(h.history['loss'], label='Train')
    plt.plot(h.history['val_loss'], label='Validation')
    plt.title(f'Loss - {name}')
    plt.legend()
    plt.show()

# ====== KONFIGURASI TERBAIK (P4.3) ======
best = df_results.sort_values(by='Val_Accuracy', ascending=False).iloc[0]
print("KONFIGURASI TERBAIK:")
print(best)
print("""
Konfigurasi ini dipilih karena memberikan akurasi validasi tertinggi
dan loss terendah. Aktivasi ReLU mempercepat konvergensi dan learning rate
yang sesuai menjaga stabilitas proses training.
""")

# ==========================================
# EVALUASI MODEL AKHIR (TEST SET)
# ==========================================

from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error, mean_absolute_error
import seaborn as sns

# 1️⃣ Bangun ulang MODEL TERBAIK
best_activation = 'relu'
best_lr = 0.0001   # sesuai hasil df_results (Val_Accuracy tertinggi)

final_model = build_model(best_activation, best_lr)

print("\nTraining ulang MODEL TERBAIK pada data training...")
final_history = final_model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# 2️⃣ Evaluasi pada TEST SET
test_loss, test_accuracy = final_model.evaluate(X_test, y_test, verbose=0)

print("\n=== PERFORMA MODEL AKHIR PADA TEST SET ===")
print(f"Test Accuracy : {test_accuracy:.4f}")
print(f"Test Loss     : {test_loss:.4f}")

# 3️⃣ Prediksi untuk evaluasi lanjutan
y_pred_prob = final_model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

# 4️⃣ Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Model Akhir")
plt.show()

# 5️⃣ Classification Report
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# 6️⃣ MAE & MSE
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("=== METRIK TAMBAHAN ===")
print(f"MSE : {mse:.4f}")
print(f"MAE : {mae:.4f}")

# 7️⃣ Analisis Overfitting / Underfitting
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(final_history.history['accuracy'], label='Train Accuracy')
plt.plot(final_history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Final Model - Accuracy Curve")
plt.legend()

plt.subplot(1,2,2)
plt.plot(final_history.history['loss'], label='Train Loss')
plt.plot(final_history.history['val_loss'], label='Validation Loss')
plt.title("Final Model - Loss Curve")
plt.legend()

plt.show()

print("""
=== ANALISIS MODEL AKHIR ===
Jika kurva training dan validation accuracy saling mendekati
dan loss validation tidak meningkat tajam, maka model tidak
mengalami overfitting. Apabila kedua akurasi rendah, maka
terjadi underfitting. Berdasarkan grafik di atas, model akhir
menunjukkan performa yang stabil dan generalisasi yang baik.
""")
```
